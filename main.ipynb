```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
url = 'https://raw.githubusercontent.com/dhruv-jani-0808/House-price-prediction/main/Housing.csv'
df = pd.read_csv(url)
```

```python
#Here, X is the DataFrame of features, and y is our target price
X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
y = df['price']
```

```python
#We are splitting the dataset into two parts: 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
#This is a visualization of Area vs. Price to understand the initial data trend
plt.scatter(df['area'], df['price'], color='blue', alpha=0.5)
plt.title('House Area vs Price')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.show()
```

```python
#Creating and training the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Predicting outputs for the test dataset
y_pred = model.predict(X_test)
```

```python
#Calculating the Mean Absolute Error (MAE) and $R^2$ score
MAE = metrics.mean_absolute_error(y_test, y_pred)
r2Score = metrics.r2_score(y_test, y_pred)
```

```python
#Visualizing actual vs. predicted house prices
print(f"Mean abosolute error : {MAE:.2f}\nR2 score : {r2Score:.4f}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted House Prices (Multiple Features)')
plt.show()
```
